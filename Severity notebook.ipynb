```python
# Import the essential libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

STEP 1: LOADING THE DATASET

```python
# Load the dataset.
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RTA Dataset.csv')

# Get the basic shape and info
# df.sample allows you to run the code multiple times so it can act a "quick preview button"
print("Dataset shape:",df.shape)
df.sample(7)
```

```python
# Finding the number of incidences per severity.
severity_column = 'Accident_severity'
print("Unique values in Severity column:")
print(df[severity_column].unique())
print("Value counts:")
print(df[severity_column].value_counts())
```

STEP 2: DATA EXPLORATION AND FEATURE ENGINEERING

```python
#Step 1: understanding the contents of the dataset, detecting outliers, filling or dropping columns as well as encoding.


#  1. Check the data types of each column
print(" Data Types of RTA Dataset:")
display(df.dtypes.value_counts())  # quick summary of how many of each type

#  2. Show columns with their data types
print(" Detailed Column Types:")
display(df.dtypes)

#  3. Check for missing values (count per column)
print("Missing Values Per Column")
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)
display(missing_values)

#  4. Percentage of missing values for better context
print("\n Percentage of Missing Values:")
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage = missing_percentage[missing_percentage > 0].sort_values(ascending=False)
display(missing_percentage)

```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```
